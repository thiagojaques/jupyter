```python
'''
Vamos começar verificando se temos o código da unidade anterior. Se você estiver executando isso no Visual Studio Code, confirme se você tem esse código em seu bloco de anotações.
'''
print('Artemis Rover Rock Scanner Iniciando')

basalt = 0
breccia = 0
highland = 0
regolith = 0
rockList = []

'''
Vamos criar um arquivo de amostra que podemos usar ao longo deste módulo. Isso conterá uma lista de rochas.

Execute esta célula de código para:

Especifique o arquivo de texto no qual escreveremos
Abra o arquivo
Escrever texto no arquivo
Feche o arquivo
'''

strPath = 'rocks.txt'
fileObject = open(strPath, 'w')
fileObject.writelines(['Carregando Rochas\n', 'basalt\n', 'breccia\n', 'highland\n', 'regolith\n', 'highland\n', 'breccia\n', 'highland\n', 'regolith\n', 'regolith\n', 'basalt\n', 'highland\n', 'basalt\n', 'breccia\n', 'breccia\n', 'regolith\n', 'breccia\n', 'highland\n', 'breccia\n', 'basalt\n',])
fileObject.close()

'''
Agora que temos algumas variáveis feitas, podemos começar a dar valores a essas variáveis. Queremos contar o número de vezes que vemos um determinado tipo de rocha no arquivo rocks.txt e adicionar esse número à variável correspondente. Por exemplo, se houver 12 rochas de basalto que Artemis salvou no arquivo de texto, queremos que a variável de basalto seja definida como 12.

Vamos começar dizendo ao computador para ler o arquivo rocks.txt . Para fazer isso, execute o código abaixo. A readlinefunção lê a primeira linha, que está apenas nos dizendo que estamos começando a ler as rochas que estão escritas no arquivo.
'''

strPath = 'rocks.txt'
fileObject = open(strPath)
line = fileObject.readline()
print(line)


'''
Agora, podemos chamar a função readlines() para colocar todas as linhas em nossa rockList variável. Também imprimiremos todas as rochas que estão no rockList e fechamos o fileObject.
'''

rockList = fileObject.readlines()

for rock in rockList:
    print(rock)

fileObject.close()

'''
A próxima etapa em nosso aplicativo é adicionar uma função que examina uma linha de texto, informa qual é o tipo de rocha e incrementa a variável correspondente. Para começar, execute esta célula de código:
'''

def countMoonRocks(rockToID):
    global basalt
    global breccia
    global highland
    global regolith

    rockToID = rockToID.lower()
    
    if('basalt' in rockToID):
        print('Encontramos basalto\n')
        basalt += 1
    elif('breccia' in rockToID):
        print('Foi encontrado breccia\n')
        breccia += 1
    elif('highland' in rockToID):
        print('Encontramos highland\n')
        highland += 1
    elif('regolith' in rockToID):
        print('Dessa vez foi encontrado regolith\n')
        regolith += 1
    
    return

'''
Agora que a função está feita, precisamos incluir uma chamada para que ela seja usada. Vamos criar um for loop para passar por cada valor em nossa lista de rock e chamar a função com cada rock.
'''
for rock in rockList:
    countMoonRocks(rock)

'''
Este código chama a função que criamos para cada linha de texto (cada rocha) que foi reportada pelo rover. A função adiciona a contagem do tipo de rocha correspondente e armazena esse número na respectiva variável.

Agora que temos todas as contagens de pedras em suas variáveis, podemos usar algum código adicional para escrever resumos sobre elas.

Vamos começar imprimindo o número de cada tipo de rocha que encontramos. Por exemplo, para imprimir o número de rochas basálticas, execute esta célula de código:
'''

print('Número de Pedras Basalto', basalt)
print('Número de Pedras Breccia', breccia)
print('Número de Pedras Highland', highland)
print('Número de Pedras Regolith', regolith)
print('O número máximo de rocha achada foi: ', max(basalt, breccia, highland, regolith))
print('E o número de rocha menos achada foi: ', min(basalt, breccia, highland, regolith))
```
